```python
def split(text):
  return text.split('Situation')[0]
def split_sit(text):
  return text.split('Situation:')[1]
```

```python
df['law description']=df['instruction'].apply(split)
df['situation']=df['instruction'].apply(split_sit)
```

```python
def split_list(x,i):
  k=x[i].split(",")[0].replace('\n',"")
  if(("not" in k) or ("No" in k)):
    return "No"
  else:
    return "Yes"
#df['law description'] = df['law description'].apply(lambda x: x.replace('\\n', ''))
```

```python
for i in range(5):
  df[f'predicted_{i+1}']=df['predicted_output'].apply(lambda x: split_list(x,i))
```
